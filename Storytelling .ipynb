# Problem:
We have been given a survey of RR Diner Coffee's 710 most loyal customers. From this we need to make a compelling decision tree model capable of predicting how much of the Hidden Farms brand from rural China will be purchased. If the amount is over 70% RR will go through with the deal.

# Data Wrangling and Cleaning:
We looked at the data's dimensions (702 rows, 9 Columns), this was not unreasonably large so, for the time being we didn't augment that. We noticed some null values in the data types, however. We needed to determine whether that would effect our ability to make a model, so we really started digging in. We tidied some of the column names to make them more specific (i.e., 'spent_week" was changed to "spent_last_week"), and adjusted the capitalization in the gender column to be consistent. Finally, we tackled those aforementioned Null values, which turned out to be in the "Decision" column, representing a "No" answer, so we changed the 1.0 and Null values to "YES" and "NO" respectively.

# EDA and Setting Up the Model:

Looking at the data we found there was a correlation between distance to the coffee shop and the desire to buy Hidden Farms coffee. Those who lived farther away would be likelier to be a nonbuyer than those who lived close by. 



We compiled all the features--that is all the columns besides the "Decision" column then we split the test and training data making sure to first det up our dummy values so as not to cause fitting issues later on. We settled on a random state of 246 and a test size of 0.25 meaning 25% of the data was in the test split.  

# Models:

We ran a series of models: Entropy (No Depth), Gini (No Depth), Entropy (Depth of 3), and Gini (Depth of 3). The first two models gave us great accuracy but they were wasteful in processing power being that they went unnecessarily deep. The Depth-limited entropy model suffered a larger accuracy loss than the depth-limited Gini model (which maintained an accuracy of 97), so the Depth-limited gini model was selected as the best of those run.

This falls short of RR's desired outcome for the deal to go through, but it might be within range of negotiation.

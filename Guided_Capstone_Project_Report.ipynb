# Big Mountain Resort 


# Context
Big Mountain Resort receives roughly 350,000 customers annually and utilizes a market-guided approach to pricing, charging a slight premium over average. We were asked to come up with alternative pricing models via utilization of any of the following parameters:

1) Increasing ticket pricing even higher based on facilities advantages unique to Big Mountain.

2) Cutting operational costs.

# Facilities Controlling Market Pricing

Models indicate a total of eight facilities as price significant: vertical drop, snow making acreage, total number of chairs, number of fast Quads, number of runs, length of longest run, number of trams, and the size of the skiable terrain. Of these, Big Mountain Resort nears the top of the resorts in all features save for vertical  drop (Fig. 3).

Given Big Mountain's current pricing of 81.00 per ticket we could justify an increase to 95.87 based on the current facilities. This falls outside the model error range of $10.39, which indicates that the price improvement is sound. The assumption of the model is that ticket pricing as provided by the data is accurate given to what consumers will pay for a given facilitiy. It would not be possible to detect over or underpricing by competing resorts given our current model/data set.

# Recommendation

Scenarios 2 and 3 are the most useful for Big Mountain Resorts to consider. If there is greater concern for the operational overhead, then Scenario 2 should be implemented over 3 as 3 will require additional snow machines and upkeep. Closing one run, per scenario 1, would not impact the ticket pricing at all while opening another run with an increased vertical drop could lead to even futher increased ticket pricing--the additional chair will cost the business 1,540,000 in operations, with 0.88 per ticket going into development. This would result in a net revenue of 13,525,471.
